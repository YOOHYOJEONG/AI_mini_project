```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import pandas as pd

#load_data
b_cancer = load_breast_cancer()
b_cancer_data = b_cancer.data
b_cancer_label = b_cancer.target

#check_data
print('data shape :', b_cancer_data.shape)
print(b_cancer_label.shape)
print('data[0] : \n', b_cancer_data[0])
print('label :', b_cancer_label)
print('target_names :', b_cancer.target_names)
print('keys : ', b_cancer.keys())

b_cancer_df = pd.DataFrame(data = b_cancer_data, columns = b_cancer.feature_names)
b_cancer_df["label"] = b_cancer.target
b_cancer_df
```

```python
print(b_cancer.DESCR)
```

```python
#train, test data separate

X_train, X_test, y_train, y_test = train_test_split(b_cancer_data, 
                                                    b_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=21)

print('X_train : ', len(X_train), ', X_test : ', len(X_test))
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
#Decision Tree Model

from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print('accuracy : ', accuracy)

print('confusion matrix : \n', confusion_matrix(y_test, y_pred))
```

```python
#Random Forest Model

from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print('accuracy : ', accuracy)

print('confusion matrix : \n', confusion_matrix(y_test, y_pred))
```

```python
#SVM Model
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print('accuracy : ', accuracy)

print('confusion matrix : \n', confusion_matrix(y_test, y_pred))
```

```python
#SGD Classifier Model

from sklearn.linear_model import SGDClassifier

sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print('accuracy : ', accuracy)

print('confusion matrix : \n', confusion_matrix(y_test, y_pred))
```

```python
#Logistic Regression Model

from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print('accuracy : ', accuracy)

print('confusion matrix : \n', confusion_matrix(y_test, y_pred))
```
